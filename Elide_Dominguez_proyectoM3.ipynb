```python
#Importamos librerias
import random
import matplotlib.pyplot as plt

#Definir funcion para la simulacion del recorrido de la canica
def simular_canica(rebotes=12):

    """Simular el recorrido que hace una canica en 12 rebotes.
    Cada rebote puede ir a la izquierda(0) o a la derecha(1)"""

    posicion = 0
    for _ in range(rebotes):
        decision = random.choice([0,1]) # 0 = izq, 1 = der
        posicion += decision
    return posicion

#Definir funcion para simular maquina
def maquina_de_galton(num_canicas=3000, rebotes = 12): #Determinar que seran 3000 canicas en 12 rebotes

    """Simula la maquina del galton con un numero determinado de canicas"""

    resultados = []
    for _ in range(num_canicas):
        contenedor = simular_canica(rebotes)
        resultados.append(contenedor)
    return resultados

#Ejecucion
resultados = maquina_de_galton()

#Graficar los resultados
plt.hist(resultados, bins=range(14), edgecolor="red", align='left')
plt.title('Simulacion de la maquina de Galton') #Titulo de la grafica
plt.xlabel('Distrubucion de canicas') #Nombre eje x
plt.ylabel('Cantidad de canicas') #Nombre eje y
plt.xticks(range(13))
plt.grid(axis='y', linestyle= '--', alpha=0.7, color='green')
plt.show()
```
